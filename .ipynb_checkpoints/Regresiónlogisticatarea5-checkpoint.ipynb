```python
#1. Realizar un análisis exploratorio de los datos para identificar relaciones entre variables, valores atípicos, tendencias, etc.
import pandas as pd
import numpy as np
import seaborn as sb
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
#Identificación de datos faltante y atípicos
#Identificación de datos faltante
plt.figure(figsize=(6,3))
sns.boxplot(x=Datos['Age'])
plt.title('Identificación de Datos Faltantes y Atípicos', fontsize=10)
```

```python
#Identificación de datos faltante 
plt.figure(figsize=(6,3))
sns.boxplot(x=Datos['Fare'])
plt.title('Identificación de Datos Faltantes y Atípicos', fontsize=10)
```

```python
# 2. Preprocesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario.
#Datos faltantes
print("Cantidad de registros con cero en Clase: ", Datos['Pclass'].isin([0]).sum())
print("Cantidad de registros con cero en SibSp: ", Datos['SibSp'].isin([0]).sum())
print("Cantidad de registros con cero en Parch: ", Datos['Parch'].isin([0]).sum())
print("Cantidad de registros con cero en Tarifa: ", Datos['Fare'].isin([0]).sum())
print("Cantidad de registros con cero en Edad: ", Datos['Age'].isin([0]).sum())
```

```python
Datos ['SibSp'] =Datos['SibSp'].replace(0,Datos['SibSp'].mean())
Datos ['Parch'] =Datos['Parch'].replace(0,Datos['Parch'].mean())
Datos ['Fare'] =Datos['Fare'].replace(0,Datos['Fare'].mean())
```

```python
print("Cantidad de registros con cero en SibSp: ", Datos['SibSp'].isin([0]).sum())
print("Cantidad de registros con cero en Parch: ", Datos['Parch'].isin([0]).sum())
print("Cantidad de registros con cero en Tarifa: ", Datos['Fare'].isin([0]).sum())
```

```python
#Identificación de datos Atípicos
plt.figure(figsize=(6,3))
sns.boxplot(x=Datos['Age'])
plt.title('Identificación de Datos Atípicos', fontsize=10)
```

```python
#Conteo de datos atípicos
nivel_min=105
nivel_max=130
Total_Atipico=((Datos['Age']<nivel_min)|(Datos['Age']>nivel_max)).sum()
print("El total de datos atípicos es {}".format(Total_Atipico))
```

```python
#Identificación de datos Atípicos
plt.figure(figsize=(6,3))
sns.boxplot(x=Datos['Fare'])
plt.title('Identificación de Datos Atípicos', fontsize=10)
```

```python
#Conteo de datos atípicos
nivel_min=200.0000
nivel_max=990.0000
Total_Atipico=((Datos['Fare']<nivel_min)|(Datos['Fare']>nivel_max)).sum()
print("El total de datos atípicos es {}".format(Total_Atipico))
```

```python
#Filtrar datos atípicos
Datos=Datos[(Datos['Fare']<nivel_max)&(Datos['Fare']>nivel_min)]
```

```python
#Validar si se removieron los datos atípicos
Total_Atipico=((Datos['Fare']<nivel_min)|(Datos['Fare']>nivel_max)).sum()
print("El total de datos atípicos es {}".format(Total_Atipico))
```
