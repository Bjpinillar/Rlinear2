```python
# 2. Preprocesar los datos limpiándolos, tratando valores faltantes y transformándolos según sea necesario.
#Datos faltantes

print("Cantidad de Registros con cero en Edad: ",Datos['Age'].isin([0]).sum())
print("Cantidad de Registros con cero en Precio de Tarifa: ",Datos['Fare'].isin([0]).sum())
```

```python
#Identificación de datos Atípicos
plt.figure(figsize=(6, 3))
sns.boxplot(x=Datos['Age'])
plt.title('Identificación de Datos Faltantes y Atípicos', fontsize=10 )
```

```python
#Conteo de datos atípicos
min=110
max=120
Total_Atipicos=((Datos['Age']<min)|(Datos['Age']>max)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
plt.figure(figsize=(6, 3))
sns.boxplot(x=Datos['Fare'])
plt.title('Identificación de Datos Faltantes y Atípicos', fontsize=10 )
```

```python
min=2.0000
max=100.0000
Total_Atipicos=((Datos['Fare']<min)|(Datos['Fare']>max)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
#Filtrar datos atípicos
Datos=Datos[(Datos['Fare']<max)&(Datos['Fare']>min)]
```

```python
#Validar si se removieron los datos atípicos
Total_Atipicos=((Datos['Fare']<min)|(Datos['Fare']>max)).sum()
print("El total de datos atípicos es {}".format(Total_Atipicos))
```

```python
#3. Seleccionar las caracteristicas más relevantes para entrenar el modelo utilizando selección de caracteristicas
#Total de personas que sobrevivieron
Datos = Datos[["Parch"]]
```

```python
Datos = Datos[['Parch']]
X = Datos.drop ('Parch',axis=1)
Y = Datos['Parch']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state= 0)
```

```python
modelo_regresion = LinearRegression() 
modelo_regresion.fit (X_train , Y_train)   
Prediccion = modelo_regresion.predict(X_test)
```
